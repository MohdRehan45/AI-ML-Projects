```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
```

```python
# Load dataset
df = pd.read_csv("language.csv")
df.head(5)
```

```python
#Create a CountVectorizer object

vectorizer = CountVectorizer()

# Sample text data

data = ["love data science", "love machine learning"]

# Fit and transform the data

vectorized_data = vectorizer.fit_transform(data)

#Get the vocabulary (unique words)

print(vectorizer.get_feature_names_out())

#Convert the result to an array

print(vectorized_data.toarray())
```

```python
#Sample emails

emails = ["I love learning", "Click here to win money", "Learning is fun", "Get the money"]
labels = [0, 1, 0, 1] # 0 means not spam, 1 means spam

#Step 1: Convert text into numbers (word counts)

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(emails)

#Step 2: Train the MultinomialNB model

model = MultinomialNB()
model.fit(X, labels)

#Step 3: Predict if a new email is spam or not

new_email = ["Win money now!"]

new_email_vector = vectorizer.transform(new_email)

prediction = model.predict(new_email_vector)

print("Prediction:", prediction)
# 1 means spam
```

```python
# 📌 Cell 4: Explore class distribution
df["language"].value_counts()
```

```python
# 📌 Cell 5: Prepare features and labels
x = np.array(df["Text"])
y = np.array(df["language"])
print(x)
```

```python
print(y)
```

```python
# 📌 Cell 6: Convert text to numerical features (Bag of Words)
cv = CountVectorizer()
X = cv.fit_transform(x)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# 📌 Cell 8: Train Naive Bayes model
model = MultinomialNB()
model.fit(X_train, y_train)
model.score(X_test,y_test)
```

```python
# 📌 Cell 10: Predict on user input
user = input("Enter a Text: ")
data = cv.transform([user]).toarray()
output = model.predict(data)
print(output)
```
